## get_Trials()

```python
def get_Trials(ratname, sessid):
    """
    Get Thomas' Trials structure from the Matlab function PB_process_data

    Parameters
    ----------
    ratname : str
        Name of the rat

    sessid : int
        Session ID   
    """
    try:
        str = """T = PB_process_data('%s', sessids=%d); T.sess_date=string(T.sess_date);""" % (ratname, sessid)
        str = str + "T.reward_type = string(T.reward_type);"
        eng.eval(str, nargout=0)
    except:
        print(f"Error with {ratname} and {sessid} -- is the Matlab engine running?")
        return None
    Trials = matArrays2Numpy(eng.workspace['T'])

    # Let's squeeze any numpy arrays (i.e. remove singleton dimensions)
    for (k, v) in Trials.items():
        if isinstance(v, np.ndarray):
            Trials[k] = v.squeeze()
        # if an entry in Trials is a list, let's try and squeeze the elements of the list
        elif isinstance(v, list):
            if (k=="leftBups") or (k=="rightBups"):
                # For right bups and left bups we want to make sure that we have one-d numpy arrays, not scalars
                Trials[k] = list(map(lambda x: np.atleast_1d(x) if isinstance(x, float) else np.atleast_1d(x.squeeze()) if isinstance(x, np.ndarray) else x, v))
            else:
                # we only try to squeeze the elements of the list if they are numpy arrays
                Trials[k] = list(map(lambda x: x.squeeze() if isinstance(x, np.ndarray) else x, v))
    # and let's squeeze each of the entries in stateTimes
    for (k, v) in Trials["stateTimes"].items():
        if isinstance(v, np.ndarray):
            Trials["stateTimes"][k] = v.squeeze()

    return Trials
```

```python
@schema
class BehaviorTrials(dj.Manual):
    definition = """
    # session-and-trial-by-trial info on behavior. StateMachine details are in the BehaviorEvent table
    -> Sessions
    trial: int
    ---
    violated=null:     bool      # True if the subject violated the task structure (e.g. broke fixation too early)
    never_cpoked=null: bool      # True if the subject never engages, i.e., never even poked the center port
    poked_r=null:      bool      # True if the subject did not violate and responded by poking the right port. False if they responded left. NaN for all other conditions (violated, never_cpoked)
    is_hit=null:       bool      # True if the subject was correct in their response
    sides:        varchar(16)      # Side that holds the reward, typically 'r' or 'l'
    clicks_on=null:    float     # time at which click generation Poisson rates go above zero  (note what is available to the rat is stereo_click, the actual first click time)
    stereo_click=null: float     # Time of the starting initial stereo click
    leftbups:     blob      # Left click times, relative to stereo_click (so first click is always at 0)  
    rightbups:    blob      # Right click times, relative to stereo_click (so first click is always at 0)
    last_click=null:   float     # time of the actual last click, whether left or right (NaN if no clicks)
    clicks_off=null:   float     # time at which click generation Poisson rates go back to zero
    num_left=null:     int       # Number of left clicks  = leftbups.size
    num_right=null:    int       # Number of right clicks = rightbups.size
    is_frozen=null:    bool      # True if the clicks are from the frozen noise clicks set
    seed=null:         double    # Random seed used to generate the time of the clicks in this trial
    state_0=null:      float     # Official Bcontrol time of the start of the trial
    sending_trialnum=null: float # start of state sending trialnumber out from Bcontrol
    sync_flash=null:   float     # start of state in which an IR led flashes to help sync video with Bcontrol
    wait_for_cpoke=null: float   # start of state waiting for subject to poke center port (center LED goes on at this time)
    cpoke_in=null:     float     # time of center port poke in
    break=null:        float     # time of a center port poke out that broke required fixation and will lead to a violation
    cpoke_req_end=null: float    # time at which the animal is no longer required to hold center port fixation (center LED turns off at this time)
    cpoke_out=null:    float     # time at which the animal actually pulls out of the center port. If the animal pulls out without a violation (i.e., after cpoke_req_end) but then pokes back into the center port, cpoke_out will be the time of the final center poke
    spoke=null:        float     # time at which the animal first pokes a side port (always comes after cpoke_out; either right or left)
    error=null:        float     # time of the start of the error state, usually accompanied by a white noise sound, NaN if no error
    left_reward=null:  float     # time of left reward delivery, NaN if no reward
    right_reward=null: float     # time of right reward delivery, NaN if no reward
    feedback=null:     float     # time of feedback delivery, i.e, same as whichever of error, left rew, right rew was not NaN
    sound_delay_dur=null: float  # clicks_on - cpoke_in
    stim_dur=null:     float     # clicks_off - clicks_on
    response_dur=null: float     # spoke - cpoke_req_end
    movement_dur=null: float     # spoke - cpoke_out
    
    

    """
```

```python
def insert_behavior_trial(sid, replace=True, skip_duplicates=False):
    """
    Insert a single session into the BehaviorTrials table
    
    Parameters
    ----------
    sid : int
        Session ID
    
    replace : bool
        If True, replace any existing entries for this session

    skip_duplicates : bool
        If True, skip inserting this session if it already exists in the table

    Returns
    -------
    entries : list of dict
        List of dictionaries with the entries that were inserted. Each dict is one trial
        
    """
    if replace and skip_duplicates:
        raise ValueError("Can't replace and skip_duplicates at the same time")

    rat = (Sessions() & "sessid=%d" % sid).fetch("ratname")[0]
    print("********* session %d for rat %s ********* " % (sid, rat))
    print("Getting Trials structure from Matlab/bdata...")
    Trials = get_Trials(rat, sid)

    entries = []
    for t in range(Trials["violated"].size):
        entry = {}
        entry["sessid"] = sid
        entry["trial"] = t+1
        entry["violated"] = int(Trials["violated"][t])
        entry["never_cpoked"] = int(Trials["never_cpoked"][t])
        entry["poked_r"] = int(np.round(Trials["pokedR"][t])) if not np.isnan(Trials["pokedR"][t]) else None
        entry["is_hit"] = int(Trials["is_hit"][t])
        entry["sides"] = Trials["sides"][t]
        entry["clicks_on"] = Trials["stateTimes"]["clicks_on"][t]
        entry["leftbups"] = Trials["leftBups"][t]
        entry["rightbups"] = Trials["rightBups"][t]
        if entry["leftbups"].size > 0 and entry["rightbups"].size > 0:
            if np.abs(entry["leftbups"][0] - entry["rightbups"][0]).max() < 1e-6:
                entry["stereo_click"] = entry["leftbups"][0]
            else:
                    entry["stereo_click"] = np.nan
            if entry["leftbups"][-1] > entry["rightbups"][-1]:
                entry["last_click"] = entry["leftbups"][-1]
            else:
                entry["last_click"] = entry["rightbups"][-1]
        else:
            entry["stereo_click"] = np.nan
            entry["last_click"] = np.nan
        entry["clicks_off"] = Trials["stateTimes"]["clicks_off"][t]
        entry["num_left"] = entry["leftbups"].size
        entry["num_right"] = entry["rightbups"].size
        entry["is_frozen"] = int(Trials["is_frozen"][t])
        entry["seed"] = Trials["seed"][t]
        entry["state_0"] = (BehaviorEvent() & "sessid=%d" % sid & "event_name='state_0'" & "trial=%d" % (t+1) & "entry_num=1").fetch("out_time")[0]
        entry["sending_trialnum"] = Trials["stateTimes"]["sending_trialnum"][t]
        entry["sync_flash"] = Trials["stateTimes"]["sync_flash"][t]
        entry["wait_for_cpoke"] = Trials["stateTimes"]["wait_for_cpoke"][t]
        entry["cpoke_in"] = Trials["stateTimes"]["cpoke_in"][t]
        entry["break"] = Trials["stateTimes"]["break"][t]
        entry["cpoke_req_end"] = Trials["stateTimes"]["cpoke_req_end"][t]
        entry["cpoke_out"] = Trials["stateTimes"]["cpoke_out"][t]
        entry["spoke"] = Trials["stateTimes"]["spoke"][t]
        entry["error"] = Trials["stateTimes"]["error"][t]
        entry["left_reward"] = Trials["stateTimes"]["left_reward"][t]
        entry["right_reward"] = Trials["stateTimes"]["right_reward"][t]
        entry["feedback"] = Trials["stateTimes"]["feedback"][t]
        entry["sound_delay_dur"] = entry["clicks_on"] - entry["cpoke_in"]
        entry["stim_dur"] = entry["clicks_off"] - entry["clicks_on"]
        entry["response_dur"] = entry["spoke"] - entry["cpoke_req_end"]
        entry["movement_dur"] = entry["spoke"] - entry["cpoke_out"]

        # DataJoint can't take NaNs, so we need to convert them to None; the default null value will read as NaN on a fetch()
        for (k, v) in entry.items():
            try:
                if v.size>0 and np.isnan(v):
                    entry[k] = None
            except:
                pass

        entries.append(entry)

        if t % 10 == 0:
            print("      sid=%d: assembled %d / %d trials" % (sid, t, Trials["violated"].size))

    print("Inserting %d trials into BehaviorTrials..." % len(entries))
    BehaviorTrials.insert(entries, replace=replace, skip_duplicates=skip_duplicates)

    return entries
```
